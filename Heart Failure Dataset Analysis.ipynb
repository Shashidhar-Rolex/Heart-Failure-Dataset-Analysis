```python
# Locate input and output variable
In= df.iloc[:,:-1]
out= df.iloc[:,-1]

# Split the data into 80:20 ratio
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(In,out,test_size=0.2,random_state=1)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
List_Accy= []
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score, ConfusionMatrixDisplay, classification_report
from sklearn.metrics import precision_recall_curve, auc
```

```python
from sklearn.svm import SVC
from sklearn.metrics import roc_auc_score
from scikitplot.metrics import plot_roc_curve

SVC_clf = SVC(C =0.7, kernel= 'linear', random_state=1, probability= True)
SVC_clf.fit(x_train, y_train)
SVC_clf_pred = SVC_clf.predict(x_test)

List_Accy= []
SVC_clf_acc = accuracy_score(y_test, SVC_clf_pred)
List_Accy.append(100* SVC_clf_acc)
print(classification_report(y_test, SVC_clf_pred))

# confusion matrix
cm = confusion_matrix(y_test, SVC_clf_pred)

# Create a figure with subplots
fig, ax = plt.subplots(figsize=(6, 4))

# Create a heatmap
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=['No heart failure', 'Heart failure'],
            yticklabels=['No heart failure', 'Heart failure'], ax=ax)

plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.savefig('Confusion Matrix for SVM.png', format='png')

# Adjust layout to add space between title and heatmap
plt.subplots_adjust(top=0.9, bottom=0.15)
plt.show()

# Receiver Operating Characteristic (ROC) Curve
SVC_clf_predict_proba = SVC_clf.predict_proba(x_test)[::,1]
False_pr, True_pr, _ = metrics.roc_curve(y_test,  SVC_clf_predict_proba)
auc = metrics.roc_auc_score(y_test, SVC_clf_predict_proba)

plt.plot(False_pr, True_pr, color='blue', label="ROC curve (AUC = {:.2f})".format(auc))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line for random guessing
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
plt.savefig('ROC for SVM.png', format='png')
# precision recall curve

from sklearn.metrics import precision_recall_curve, auc
precision, recall, thresholds = precision_recall_curve(y_test, SVC_clf_pred)
area_under_curve = auc(recall, precision)

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='purple', label=f'PR Curve (AUC = {area_under_curve:.2f})')
plt.title('Precision-Recall Curve', fontsize=16)
plt.xlabel('Recall', fontsize=14)
plt.ylabel('Precision', fontsize=14)
plt.legend(loc='lower left', fontsize=12)
plt.grid()
plt.show()
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Define the parameter grid for Decision Tree
param_grid = {
    'max_depth': [20,25,27],
    'min_samples_split': [2, 3,4,5,6,7],
    'min_samples_leaf': [1, 2, 10,20,30,40,50,60,100],
}

# Initialize the Decision Tree classifier
decision_tree = DecisionTreeClassifier()

# Initialize the GridSearchCV object
grid_search = GridSearchCV(estimator=decision_tree, param_grid=param_grid, scoring='accuracy', cv=6)

# Fit the Grid Search to the data
grid_search.fit(x_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
best_dt_model = grid_search.best_estimator_

# Predict using the best model
y_pred = best_dt_model.predict(x_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.2f}")
print(best_params)
```
